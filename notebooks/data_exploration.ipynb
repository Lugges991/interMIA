```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/mnt/DATA/datasets/ABIDEII/ABIDEII_all_combined_drop_columns.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.style.use("grayscale")
```

```python
one = np.count_nonzero(df.DX_GROUP.values == 1)
two = np.count_nonzero(df.DX_GROUP.values == 2)
```

```python
plt.bar([1, 2], [one, two], tick_label=["ASD", "TC"])
plt.show()
```

```python
sites = np.unique(df.SITE_ID.values)

site_1 = []
site_2 = []

for site in sites:
    site_1.append(np.count_nonzero(df[df.SITE_ID==site].DX_GROUP.values == 1))
    site_2.append(np.count_nonzero(df[df.SITE_ID==site].DX_GROUP.values == 2))
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i])
```

```python
xs = np.arange(len(sites))

plt.rcParams["figure.figsize"] = (15,6)

fig, ax = plt.subplots()
width = .4

rects1 = ax.bar(xs - 0.2, site_1, width=width, label="ASD")
rects2 = ax.bar(xs + 0.2, site_2, width=width, label="TC")

ax.set_ylabel("Subjects")
ax.set_title("Number of ASD and TC Subjects per Site")
ax.set_xticks(xs, sites, rotation="vertical")
ax.legend()

ax.bar_label(rects1, padding=3, rotation="vertical")
ax.bar_label(rects2, padding=3, rotation="vertical")
plt.savefig("asd_tc_distribution.png")
plt.show()
```
